```python
#plot malicious traffic affected by ReDos
import matplotlib.pyplot as plt

file_path = "disrupted_traffic.txt"

# Initialize lists for line numbers and second column values
x = []
y = []

# Read the file and extract data
with open(file_path, 'r') as file:
    for line_num, line in enumerate(file, start=1):
        columns = line.strip().split(',')  # Split the line into columns
        if len(columns) > 1:
            x.append(line_num) 
            y.append(float(columns[1]))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Graph of disrupted traffic while ReDos attack running")
plt.xlabel("Request number")
plt.ylabel("Response time")
plt.ylim(0,100)
plt.grid(True)
plt.show()
```
